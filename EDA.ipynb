```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading Dataset
df=pd.read_csv("Global_Superstore_Cleaned.csv")
df
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
# Converting date datatype from object to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%y')

```

```python
df['Proft_margin_%'] = df['Proft_margin_%'].str.replace('%', '', regex=False).astype(float)
```

```python
df.info()
```

```python
# Checking null Values
df.isnull().sum()
```

```python
# Sales by category bar chart
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values()

sales_by_category.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.savefig("sales_by_category")
plt.show()
```

```python
# monthly sales trend line chart
monthly_sales = df.groupby(['Year', 'Month_no'])['Sales'].sum()

monthly_sales.plot(figsize=(12,5))

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.savefig("monthly_sales_trend")
plt.show()
```

```python
# top 10 products by sales bar chart
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(12,5))

plt.title('Top 10 Products by Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.savefig("top 10 products")
plt.show()
```

```python
# sales by market
market_sales=df.groupby('Market')['Sales'].sum().sort_values(ascending=False)
market_sales.plot(kind='bar')
plt.title('Sales_by_market')
plt.xlabel('Market')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.savefig("market_sales")
plt.show()
```

```python
# profit distribtution histogram
plt.hist(df['Profit'], bins=100)

plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.savefig("profit_distribtution")
plt.show()
```
